```python
import os

import matplotlib.pyplot as plt
import matplotlib
import numpy as np

import tensorflow as tf
assert tf.__version__.startswith('2')

from tflite_model_maker import model_spec
from tflite_model_maker import image_classifier
from tflite_model_maker.config import ExportFormat
from tflite_model_maker.config import QuantizationConfig
from tflite_model_maker.image_classifier import DataLoader

import matplotlib.pyplot as plt
```

```python
#image_path = '/content/drive/MyDrive/testimg/Augmented'
image_path = '/content/drive/MyDrive/코딩관련/colab/cnn/Augmented'

#image_path = tf.keras.utils.get_file(
     #'/content/drive/MyDrive/testimg/Augmented/dried',
     # '/content/drive/MyDrive/testimg/Augmented/dried',
     # extract=True)
#image_path = os.path.join(os.path.dirname(image_path), 'flower_photos')
```

```python
data = DataLoader.from_folder(image_path)
train_data, test_data = data.split(0.9)
```

```python
model = image_classifier.create(train_data, model_spec=model_spec.get('mobilenet_v2'),epochs=20,batch_size=128)
```

```python
train_loss, train_accuracy = model.evaluate(train_data)
```

```python
model = image_classifier.create(test_data,epochs=20,batch_size=64)
```

```python
test_loss, test_accuracy = model.evaluate(test_data)
```
